```
Y_test = np.array(df_test.ALSFRS_slope)
```

```
df_test = np.array(df_test.drop(['ALSFRS_slope'],axis = 1))
```

```
Y_train = np.array(df_train.ALSFRS_slope)
df_train = np.array(df_train.drop(['ALSFRS_slope'],axis =1))
```

```
kmeans = KMeans(n_clusters= 2)
```

```
kmeans.fit(df_train)
```

```
correct = 0
for i in range(len(df_test)):
    predict_me = np.array(df_test[i].astype(float))
    predict_me = predict_me.reshape(-1, len(predict_me))
    prediction = kmeans.predict(predict_me)
    if prediction[0] == Y_test[i]:
        correct += 1

print(correct/len(X))
```
